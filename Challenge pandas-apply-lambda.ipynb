Challenge 1:
We want to create bins of movies according to the number of votes they've received. For that matter, we will create a new column named 'bin' which will tag every movie as follow:

From 0 to 999 ==> 1
From 1000 to 9999 ==> 2
From 10000 to 99999 ==> 3
From 100000 to 999999 ==> 4
More than 1000000 ==> 5

```python
def movie_votes(votes):
    if (votes <= 999):
        return 1
    elif (votes <= 9999):
        return 2
    elif (votes <= 99999):
        return 3
    elif (votes <= 999999):
        return 4
    else:
        return 5

    
df['bin'] = df.apply(lambda x: movie_votes(x["Votes"]), axis=1)
df.head()
```

Challenge 2:
Using axis concept
We want to know how much is the revenue per minute for every movie.

```python
def revenue_minute(revenue, mins):
    return revenue/mins

df['Revenue per min'] = df.apply(lambda x: revenue_minute(x['Revenue (Millions)'], x['Runtime (Minutes)']), axis=1)
df.head()
```

Challenge 3:
Using the lambda
We want to create a new rating where we add 1 point if the genre is thriller but subtract 1 point if the genre is comedy.

```python
def new_rating(rating, genre):
    if ("Thriller" in genre):
        return rating + 1
    if ("Comedy" in genre):
        return rating - 1
    return rating
        

df['New Rating'] = df.apply(lambda x: new_rating(x['Rating'], x['Genre']), axis=1)
df.head()
```

Challenge 4:
Now the real stuff
We want to know if the sum of the ASCII value of every character of the movie title divided by the number of votes retrieve a prime number...remember, prime numbers are integers.

```python
def is_prime(title, votes):
    x = 0
    for i in title:
        x += ord(i)
    p = x / votes
    if (not p.is_integer()):
        return False
    
    print("x is " + str(x))
    print("p is " + str(p))
    
    j = 2
    
    while (j < p / 2):
        if (p % j == 0):
            return False
        j += 1
    return True

print(is_prime("J", 2))
print(is_prime("a", 2))

df['Is prime?'] = df.apply(lambda x: is_prime(x['Title'], x['Votes']), axis=1)
```

Challenge 6:
Freaky bonus
We want to know which movies might have hidden paterns in their description. A way to know that is finding those movies which the sum of all numeric values of the string description hash (SHA256) are between their revenue and their number of votes.

```python
import hashlib

def freaky_bonus(description, revenue, votes):
    h = hashlib.sha256(description.encode('UTF-8'))
    #print(h.hexdigest())
    x = 0
    if (pd.isna(revenue)):
        return False
    for i in str(h.hexdigest()):
        if i.isdigit():
            x += int(i)
    return x in range(int(votes), int(revenue))

print(freaky_bonus("asds", 2, 4))
```

```python
df['Bonus'] = df.apply(lambda x: freaky_bonus(x['Description'], x['Revenue (Millions)'], x['Votes']), axis=1)
```
